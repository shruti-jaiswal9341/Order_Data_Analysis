# Orders Data Analysis Project

## Problem Statement

In a competitive market, companies often struggle to understand which factors most significantly impact their sales performance. They face challenges in identifying the top-performing products, regions, and periods, as well as pinpointing areas with declining sales or missed opportunities. Without a clear analysis, businesses cannot optimize their marketing strategies, product offerings, or resource allocation effectively. Additionally, understanding the monthly and yearly sales growth, particularly comparing year-on-year changes, remains a critical need for continuous business improvement.

## Objective

1.	Evaluate sales performance across different time periods (e.g., month-over-month, year-over-year).
2.	Identify the top-performing products, subcategories, and regions.
3.	Analyse sales trends and patterns to forecast future performance.
4.	Determine the factors contributing to sales growth and profitability.
5.	Highlight areas of sales decline or underperformance to develop targeted strategies for improvement.

```python
# na_values is used  for convert the unused entity of a column in null value or nan
df = pd.read_csv("orders.csv",na_values=['Not Available','unknown'])
df['Ship Mode'] . unique()
```

```python
# rename columns names ...make them Lower case and replace space with underscore
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', '_')
df.columns
```

```python
df.head(5)
```

```python
#derive new columns discount.
df['discount'] = df['list_price'] * df['discount_percent']/100
df['discount']
```

```python
#calculatre the sales price and profit
df['sales_price']= df['list_price']-df['discount']
df['profit'] = df['sales_price'] - df['cost_price']
df
```

```python
df.dtypes
```

```python
#convert order date from object data type to datetime
df['order_date'] = pd.to_datetime(df['order_date'],format ="%Y-%m-%d")
df.dtypes
```

```python
#drop cost_price ,list_price,discount_percent columns
df.drop(columns=['cost_price' , 'list_price' , 'discount_percent'],inplace=True)
```
